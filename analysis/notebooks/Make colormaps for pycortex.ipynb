```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
map = np.zeros((256, 256, 4))
x, y = np.meshgrid(np.linspace(0, 1., 256), np.linspace(1., 0, 256))

for (i, j), e in np.ndenumerate(x):
    map[i, j, :3] = sns.external.husl.husl_to_rgb(x[i, j] * 360, y[i, j] * 100, 70 + (1-y[i, j]) * 30)
    
plt.imshow(map[..., :3])
```

```python
import PIL
```

```python
map *= 255
alpha = np.tile(np.linspace(4, 0, 256, endpoint=False)[:, np.newaxis], (1, 256))
alpha = np.clip(alpha, 0, 1) * 255


map[:, :, 3] = alpha

map = map.astype(np.uint8)

im = PIL.Image.fromarray(map, mode='RGBA')
im.save('/Users/gilles/miniconda3/envs/pycortex/share/pycortex/colormaps/test5_alpha.png')
im
```
